# **k-Means algorithm, Clustering**

---



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mouse.csv', header=None)
data = df.values
```

```python
data
```

```python
plt.scatter(data[:,0], data[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

## **1) n_clusters=2, max_iter=50**

```python
d1 = data.copy()
```

```python
kmeans = KMeans(n_clusters=2, max_iter=50)
kmeans.fit(d1)
```

```python
# identify the 10 closest points
distances = kmeans.transform(d1)
```

```python
distances
```

```python
# argsort returns an array of indexes which will sort the array
# in ascending order. Reverse it with [::-1]
sorted_idx0 = np.argsort(distances[:,0])[::-1][:10]
sorted_idx1 = np.argsort(distances[:,1])[::-1][:10]
```

```python
plt.scatter(d1[:,0], d1[:,1], c=kmeans.labels_)
plt.scatter(d1[sorted_idx0],
           d1[sorted_idx1],
           label='Outliers', c='black')
plt.show()
```
