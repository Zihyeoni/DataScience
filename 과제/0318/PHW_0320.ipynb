```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([[3., '?', 2., 5.], ['*', 4., 5., 6.], ['+', 3., 2., '&'], [5., '?', 7., '!']])
```

```python
df
```

```python
df.replace(('?', '*', '+', '&', '!'), np.nan, inplace=True)
```

```python
df
```

```python
df.isna().any()
```

```python
df.isna().sum()
```

```python
df_1 = df.dropna(axis=0, how='any')
df_1
```

```python
df_2 = df.dropna(axis=0, how='all')
df_2
```

```python
df_3 = df.dropna(axis=0, thresh=1)
df_3
```

```python
df_4 = df.dropna(axis=0, thresh=2)
df_4
```

```python
df_5 = df.fillna(100)
df_5
```

```python
df.describe()
```

```python
df_6 = df.fillna(df.mean())
df_6
```

```python
df_7 = df.fillna(df.median())
df_7
```

```python
df_8 = df.fillna(axis=0, method='ffill')
df_8
```

```python
df_9 = df.fillna(axis=0, method='bfill')
df_9
```
