# **Ex-1. Compute BMI**
### BMI = weight(kg) / (height(m) * height(m))
             
weight: 40.0 ~ 90.0 (kg)

height: 140 ~ 200 (cm)

---



```python
import numpy as np
```

```python
np.random.seed(0)
```

```python
wt = (90.0 - 40.0) * np.random.random_sample(100) + 40.0
ht = np.random.randint(140, 200, 100)
bmi = np.zeros(100)
```

```python
bmi = wt / ((ht*0.01) * (ht*0.01))
```

```python
print("The first 10 elements of the BMI array:")
for i in range(10):
    print(bmi[i])
```

# **Ex-2. MatPlotLib**
image.png

```python
import matplotlib.pyplot as plt
```

```python
bmi_level = ['underweight', 'healthy', 'overweight', 'obese']
num_students = [0, 0, 0, 0]
```

```python
i = 0
while i < 100:
  if bmi[i] < 18.5: # Underweight
    num_students[0] += 1
  elif 18.5 <= bmi[i] < 25.0: # Healthy
    num_students[1] += 1
  elif 25.0 <= bmi[i] < 30.0: # Overweight
    num_students[2] += 1
  elif bmi[i] >= 30.0: # Obese
    num_students[3] += 1
  i += 1
```

```python
num_students
```

```python
plt.bar(bmi_level, num_students)
plt.title("BMI bar chart")
plt.xlabel("BMI level")
plt.ylabel("Students")
plt.show()
```

```python
plt.hist(bmi, bins=4)
plt.title("BMI histogram")
plt.xlabel("BMI")
plt.ylabel("Students")
plt.show()
```

```python
plt.hist(bmi, bins=[10, 18.5, 25.0, 30.0, 45.0])
plt.title("BMI histogram (for each bmi level)")
plt.xticks([10, 18.5, 25.0, 30.0, 45.0])
plt.xlabel("BMI")
plt.ylabel("Students")
plt.show()
```

```python
plt.pie(num_students, labels=bmi_level, autopct='%1.1f%%')
plt.title("BMI pie chart")
plt.show()
```

```python
plt.scatter(ht, wt)
plt.title("BMI scatter plot")
plt.xlabel("Height(cm)")
plt.ylabel("Weight(kg)")
plt.show()
```
