## clients

```python
es = ft.EntitySet(id='clients')
```

```python
es = es.add_dataframe(
    dataframe_name='clients',
    dataframe=clients,
    index='client_id',
    time_index='joined'
)
```

## payments

```python
from woodwork.logical_types import Categorical
```

```python
es = es.add_dataframe(
    dataframe_name='payments',
    dataframe=payments,
    logical_types={'missed': Categorical},
    make_index=True,
    index='payment_id',
    time_index='payment_date'
)
```

# Aggregation

```python
stats = loans.groupby('client_id')['loan_amount'].agg('sum')
```

```python
stats = stats.to_frame()
```

```python
stats.columns = ["total_loan_amount"]
```

```python
print(stats)
```

```python
stats = clients.merge(stats, left_on='client_id', right_index=True, how='left')
```
